# Model:#

Consider a haploid population of size $N,$ which is composed of two alleles $A$ and $a$ present in 1 and $N - 1 $copies respectively. The fitnesses of $A$ and $a$ are $1 + s$ and 1 respectively, where $s > 0$. 


At each discrete time step, one individual is chosen proportional to its fitness from the population.
It produces $U$ number of progenies, drawn from the offspring distribution $P_{U}$. In order to keep the population size constant in every time step, $U$ individuals are selected to die randomly from other N âˆ’ 1 individuals.


![myimage-alt-tag](1.png)


The offspring distribution ($P_{U}(u)$) is
             \begin{equation}
              P_{U}(u) = \frac{A}{u^{\gamma}}, 
              \end{equation}
               where $A$ is the normalization constant.
 
 
*In the Moran process, $U$ can take only one value i.e., 2, while in this model, $U$ is a random variable that can take any value ranging from 2 to $N$.*


**Important: $\gamma$ is the parameter of the model. When $\gamma >> 3, $ this model reduces to the standard Moran process (see Fig 1-3).**



## Aim: to calculate the probability with which the allele $A$ will take over the population, i.e., the fixation probability of $A$.##

## Some results ##

```python
import matplotlib.pyplot as plt
import numpy

x_p = list(numpy.loadtxt("fixation_pop_size_50_s_0.5.txt", comments = '#', usecols = (0)))

data_50 = list(numpy.loadtxt("fixation_pop_size_50_s_0.5.txt", comments = '#', usecols = (1)))

plt.xlim([1.8, 7.0])
plt.ylim([0.01, 0.5])

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Model parameter (gamma)')
plt.ylabel('Fixation probability of beneficial allele A')
plt.title('N = 50, s = 0.5')

plt.hlines(0.336, 1.8, 7.0, linestyles = 'dotted' , label = 'Standard Moran process')
plt.scatter(x_p, data_50, c = 'blue', marker = 's', label = 'Modified Moran Process')

plt.legend()
```

```python
data_100 = list(numpy.loadtxt("fixation_pop_size_100_s_0.5.txt", comments = '#', usecols = (1)))

plt.xlim([1.8, 7.0])
plt.ylim([0.01, 0.5])

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Model parameter (gamma)')
plt.ylabel('Fixation probability of beneficial allele A')
plt.title('N = 100, s = 0.5')

plt.hlines(0.335, 1.8, 7.0, linestyles = 'dotted' , label = 'Standard Moran process')
plt.scatter(x_p, data_100, c = 'red', marker = 'v', label = 'Modified Moran Process')
plt.legend()
```

```python
data_500 = list(numpy.loadtxt("fixation_pop_size_500_s_0.5.txt", comments = '#', usecols = (1)))
plt.xlim([1.8, 7.0])
plt.ylim([0.01, 0.4])

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Model parameter (gamma)')
plt.ylabel('Fixation probability of beneficial allele A')
plt.title('s = 0.5, Modified Moran Process')

plt.scatter(x_p, data_50, c = 'blue', marker = 's', label = 'N = 50')
plt.scatter(x_p, data_100, c = 'red', marker = 'v', label = 'N = 100')
plt.scatter(x_p, data_500, marker = 'D', label = 'N = 500')

plt.legend()
```
